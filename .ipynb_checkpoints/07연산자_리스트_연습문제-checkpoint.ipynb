## 26. 리스트 잘라내기(슬라이싱)
다음과 같이 코드 실행 결과가 나오도록 코드의 빈칸을 완성한 전체 코드를 작성하세요.<br>
실행결과
```
b : [1, 2, 3, 4, 5]
c : ['a', 'b', 'c', 'd', 'e']
d : [5, 'a']
e : [1, 3, 5, 'b', 'd']
f : [2, 4, 'a', 'c', 'e']
```

슬라이싱 [시작인덱스:끝인덱스+1:step] step = 1

````python
a = [1,2, 3, 4, 5, 'a', 'b', 'c', 'd', 'e']
print("b:", a[:5])
print("c:", a[5:])
print("d:", a[4:6])
print("e:", a[0::2])
print("f:", a[1::2])
````

## 31. 여러 개의 값 입력받기(input + split)
세 개의 숫자를 입력받아 가장 작은 숫자를 출력하는 코드를 작성하세요.<br>
입력
```
10 2 123
```
출력
```
2
```

* 숫자입력받기
* 문자를 숫자로 변환
* if문으로 가장 작은 숫자 찾기
* 결과출력

````python
num1, num2, num3 = input().split()
num1 =int(num1)
num2 =int(num2)
num3 =int(num3)
if num1 < num2 < num3:
    print(num1)
elif num1 < num3 < num2:
    print(num1)
elif num2 < num1 < num3:
    print(num2)
elif num2 < num3 < num1:
    print(num2)
elif num3 < num1 < num2:
    print(num3)
elif num3 < num2 < num1:
    print(num3)
````
