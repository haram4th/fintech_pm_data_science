# yes24best 데이터 수집 basic

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
url = "https://www.yes24.com/Product/Category/BestSeller"
payload = dict(categoryNumber='001', pageNumber=1, pageSize=24)
r = requests.get(url, params=payload)
print(r.status_code)
print(r.text)
soup = bs(r.text, 'lxml')
```

```python
def text_clean(text):
    if type(text) == list:
        cleaned = [item.replace("\r", "").replace("\n", "").replace("\t", "").strip() for item in text]
        return ", ".join(cleaned) 
    else:
        return text.replace("\r", "").replace("\n", "").replace("\t", "").strip()
```

```python
def text_clean2(text):
    if type(text) == list:
        cleaned =[]
        for item in text:
            temp =item.replace("\r", "").replace("\n", "").replace("\t", "").strip()
            cleaned.append(temp)
        return ", ".join(cleaned)
    else:
        return text.replace("\r", "").replace("\n", "").replace("\t", "").strip()
```

```python
book_list = soup.select("#yesBestList > li")
```

```python
book_list_result = {}
for book in book_list:
    kind = book.select_one(".gd_res").string # 종류
    title = book.select_one(".gd_name").string # 책제목
    book_link = "https://www.yes24.com" + book.select_one(".gd_name")['href'] # 도서상세링크  book_link
    authors = text_clean((book.select_one(".authPub.info_auth").text).split("/")[:-1] if '역' in (book.select_one(".authPub.info_auth").text).split("/")[-1] else book.select_one(".authPub.info_auth").text) # 저자명
    trans = text_clean((book.select_one(".authPub.info_auth").text).split("/")[-1] if '역' in (book.select_one(".authPub.info_auth").text).split("/")[-1] else "") # 역자명
    publisher = book.select_one(".authPub.info_pub > a").string # 출판사명
    pub_date = book.select_one(".authPub.info_date").string # 출간일
    price = book.select_one(".info_row.info_price > strong > em").string  # 가격
    point = (book.select_one(".info_row.info_price > .yPoint").text).replace("포인트적립",  "") # 포인트
    selling_point = (book.select_one("span.saleNum").text).replace("판매지수",  "").strip() # 판매지수
    n_reviews = book.select_one(".txC_blue").text if book.select_one(".txC_blue") != None else "" # 회원리뷰 수
    rating = book.select_one(".rating_grade > em").text if book.select_one(".rating_grade > em") != None  else ""# 평점
#     print(kind, title, book_link, authors, trans, publisher, pub_date, price, point, selling_point, n_reviews, rating)
    values = [kind, title, book_link, authors, trans, publisher, pub_date, price, point, selling_point, n_reviews, rating]
    keys = ['kind', 'title', 'book_link', 'authors', 'trans', 'publisher', 'pub_date', 'price', 'point', 'selling_point', 'n_reviews', 'rating']    
    for key, value in zip(keys, values):
        book_list_result.setdefault(key, []).append(value)

    
final_result = pd.DataFrame(book_list_result)
final_result.to_csv("yes24best_list.csv", encoding="cp949", index=False)
    
```
